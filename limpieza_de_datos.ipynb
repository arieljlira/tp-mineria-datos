```python

class TIPO_ACCESO():
    ASSET = "asset"

    BUSQUEDA = "search"
    BROWSE = "search"
    VARIOS_VALIDOS = "valid"
    DSO = "valid"
    DESCARGA = "download"

    BOT_RESOURCE = "bot-resource"
    OPEN_SEARCH = "external-agent"
    OAI = "external-agent"
    SESSION = "session"
    BACKEND = "backend"
    LEGACY = "legacy"
    IGNORE = "ignore"
    UNKWOWN = "unknown"

# tipificar accesos en base a la ruta accedida
# param path request_uri
#
def tipificar_acceso_por_ruta(path):

    # Esto se hace a partir del conocimiento de la aplicación
    # el orden es necesario
    patrones_dspace = {
        '/blog/.*': TIPO_ACCESO.IGNORE,
        '/themes/.*\.(gif|jpg|png|css|js)': TIPO_ACCESO.ASSET,
        '/static/.*\.(gif|jpg|png|css|js)': TIPO_ACCESO.ASSET,
        '/images/.*\.(gif|jpg|png|css|js)': TIPO_ACCESO.ASSET,
        '/loadjQuery.js': TIPO_ACCESO.ASSET,
        '/favicon.ico': TIPO_ACCESO.ASSET,
        '/apple-touch.*.png': TIPO_ACCESO.ASSET,
        '/open-search/.*': TIPO_ACCESO.OPEN_SEARCH,
        '/feed/.*': TIPO_ACCESO.OPEN_SEARCH,
        '/sitemap.*': TIPO_ACCESO.BOT_RESOURCE,
        '/robots.txt': TIPO_ACCESO.BOT_RESOURCE,
        '/oai/.*': TIPO_ACCESO.DESCARGA,
        '/(password-login|login|logout|forgot|register)': TIPO_ACCESO.SESSION,
        '/pages/.*': TIPO_ACCESO.VARIOS_VALIDOS,
        '/community-list': TIPO_ACCESO.VARIOS_VALIDOS,

        '/.*search-filter.*': TIPO_ACCESO.BUSQUEDA,
        '/.*discover.*': TIPO_ACCESO.BUSQUEDA,
        '/.*browse.*': TIPO_ACCESO.BROWSE,
        '/bitstream/.*': TIPO_ACCESO.DESCARGA,
        '/handle/.*': TIPO_ACCESO.DSO,

        '/?id=ARG': TIPO_ACCESO.LEGACY, #etsa no va a matchear
        '/index.php/?id=ARG': TIPO_ACCESO.LEGACY,
        '/recorrido/': TIPO_ACCESO.LEGACY,
        '/moz-search-plugin.xml': TIPO_ACCESO.IGNORE,
        '/norm.php': TIPO_ACCESO.IGNORE,

        # considerar accesos al backend con sesion activa  validar siguientes
        #'(admin|profile|choices)': TIPO_ACCESO. BACKEND

        '/$': TIPO_ACCESO.VARIOS_VALIDOS,

    }
    for patron, valor in patrones_dspace.items():
        if re.match(patron,path):
            return valor

    #No encontró nada
    return TIPO_ACCESO.UNKWOWN

accesos['tipo'] = accesos['recurso'].apply(tipificar_acceso_por_ruta)

#print ("Se eliminan los accesos ignorables: ", (accesos['tipo'] == TIPO_ACCESO.IGNORE).sum())
print ("Se clasificaron los accesos como sigue: ", (accesos['tipo']).value_counts())

#  busca los que tienen tipo ignore y los saca de la lista de accesos
accesos.drop(accesos[accesos['tipo'] == TIPO_ACCESO.IGNORE].index, inplace=True)
```

```python

# Revisa un request y determina si es un intento de escaneo de aplicacion o de vulnerabilidad
# param path request_uri
#
def isScanRequest(path):

    patrones_scan = {
        '/wp-.*': True,
        '/xmlrpc.php': True,
        # considerar accesos a archivos ocultos
        '.*/\.\..*': True,
        #'.*\.php': True
    }

    for patron, valor in patrones_scan.items():
        if re.match(patron,path):
            return True

    #No encontró nada
    return False

accesos['isScan'] = accesos['recurso'].apply(isScanRequest)
print ("Se clasificaron los accesos como isScan: ", (accesos['isScan']).value_counts())
```

```python
# crear columna hasReferer
accesos['hasReferer'] = accesos['referer'].notna()

#normaliza a 200, 300, 400, 500)
accesos["codRespuesta"] = accesos["codRespuesta"].map(lambda statusCode: (np.floor(statusCode / 100) * 100).astype(int), na_action='ignore')

# Transforma codRespuesta a OneHotEncoding
one_hot = pd.get_dummies(accesos["codRespuesta"], prefix="status")
new_accesos = accesos.join(one_hot)
# print ("voy a agregar las siguientes dummies para status code" , one_hot)

# Transforma method a OneHotEncoding
one_hot = pd.get_dummies(accesos["metodo"], prefix="method")
new_accesos = new_accesos.join(one_hot)
# print ("voy a agregar las siguientes dummies para status code" , one_hot)

# Transforma tipo a OneHotEncoding
one_hot = pd.get_dummies(accesos["tipo"], prefix="access_type")
new_accesos = new_accesos.join(one_hot)

# print ("voy a agregar las siguientes dummies para status code" , one_hot)

# drop de columnas innecesarias
new_accesos= new_accesos.drop(['referer', 'dia','hora','minuto','segundo', 'codRespuesta', 'metodo'], axis=1)

new_accesos.sample(10)
```

```python
# Encuentra y retorna la sesión más larga encontrada en el intervalo especificado
# PARAMS df, Nombre de columna con fecha en el df, ventana de tiempo
# RETURNS inicio, fin,  duración
# Ejemplo de uso:
# inicio, fin, duracion = sesion_mas_larga(df, 'ColumnaDeTiempo', pd.Timedelta(minutes=30))
# print("La sesión más larga va desde", inicio, "hasta", fin, "con una duración de", duracion)
def sesion_mas_larga(df, columna_tiempo, session_time):

    df.loc[:, columna_tiempo] = pd.to_datetime(df[columna_tiempo])

    # Ordena el DataFrame por la columna de tiempo
    df_sorted = df.sort_values(by=columna_tiempo)

    # Inicializa variables para mantener el inicio y fin de la sesión más larga
    inicio_sesion = None
    fin_sesion = None
    duracion_sesion_max = pd.Timedelta(0)

    # Itera sobre las filas del DataFrame
    for index, row in df_sorted.iterrows():
        if inicio_sesion is None:
            # Si es la primera ocurrencia, establece el inicio de la sesión
            inicio_sesion = row[columna_tiempo]
            fin_sesion = row[columna_tiempo]
            inicio_sesion_max = row[columna_tiempo]
            fin_sesion_max = row[columna_tiempo]
        else:
            # Comprueba si la diferencia de tiempo con la ocurrencia anterior es mayor que session_time
            if row[columna_tiempo] - fin_sesion > session_time:
                # Si es mayor, actualiza la duración de la sesión más larga y el inicio y fin de la sesión
                duracion_sesion = fin_sesion - inicio_sesion
                if duracion_sesion > duracion_sesion_max:
                    duracion_sesion_max = duracion_sesion
                    inicio_sesion_max = inicio_sesion
                    fin_sesion_max = fin_sesion
                # Establece el inicio de una nueva sesión
                inicio_sesion = row[columna_tiempo]
            # Actualiza el fin de la sesión actual
            fin_sesion = row[columna_tiempo]

    # Si hay alguna sesión encontrada, actualiza la duración de la sesión más larga
    if inicio_sesion is not None:
        duracion_sesion = fin_sesion - inicio_sesion
        if duracion_sesion > duracion_sesion_max:
            duracion_sesion_max = duracion_sesion
            inicio_sesion_max = inicio_sesion
            fin_sesion_max = fin_sesion

    # Devuelve el inicio y fin de la sesión más larga junto con su duración
    return inicio_sesion_max, fin_sesion_max, duracion_sesion_max

import struct

def ip2long(ip):
    """
    Converts an IPv4 address from dotted-decimal notation to a long integer.

    Args:
        ip: The IPv4 address in dotted-decimal notation (e.g., "192.168.1.1").

    Returns:
        The long integer representation of the IP address.
    """

    packed_ip = struct.pack('!BBBB', *map(int, ip.split('.')))
    return struct.unpack('!L', packed_ip)[0]
```
